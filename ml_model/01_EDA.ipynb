```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Settings
pd.set_option('display.max_columns', None)
sns.set_theme(style="whitegrid")

# Load
df = pd.read_csv('../data/Churn_Modelling.csv')
df.head()
```

```python
print("Shape:", df.shape)
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())
print("\nDuplicate Rows:", df.duplicated().sum())
```

```python
# RowNumber, CustomerId, Surname have no predictive value
df.drop(columns=['RowNumber', 'CustomerId', 'Surname'], inplace=True)
print("Remaining columns:", df.columns.tolist())
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Count
df['Exited'].value_counts().plot(kind='bar', ax=axes[0], color=['steelblue', 'tomato'])
axes[0].set_title('Churn Count')
axes[0].set_xticklabels(['Stayed', 'Churned'], rotation=0)

# Percentage
df['Exited'].value_counts(normalize=True).plot(kind='pie', ax=axes[1],
    labels=['Stayed', 'Churned'], autopct='%1.1f%%', colors=['steelblue', 'tomato'])
axes[1].set_title('Churn Rate')
axes[1].set_ylabel('')

plt.tight_layout()
plt.savefig('../screenshots/churn_distribution.png', dpi=150)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Geography
geo = df.groupby('Geography')['Exited'].mean().sort_values(ascending=False)
geo.plot(kind='bar', ax=axes[0], color='steelblue')
axes[0].set_title('Churn Rate by Geography')
axes[0].set_ylabel('Churn Rate')
axes[0].set_xticklabels(geo.index, rotation=0)

# Gender
gen = df.groupby('Gender')['Exited'].mean().sort_values(ascending=False)
gen.plot(kind='bar', ax=axes[1], color='tomato')
axes[1].set_title('Churn Rate by Gender')
axes[1].set_ylabel('Churn Rate')
axes[1].set_xticklabels(gen.index, rotation=0)

plt.tight_layout()
plt.savefig('../screenshots/churn_by_category.png', dpi=150)
plt.show()
```

```python
num_cols = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']

fig, axes = plt.subplots(2, 3, figsize=(15, 8))
axes = axes.flatten()

for i, col in enumerate(num_cols):
    df[df['Exited'] == 0][col].plot(kind='hist', ax=axes[i], alpha=0.6,
                                     color='steelblue', label='Stayed', bins=30)
    df[df['Exited'] == 1][col].plot(kind='hist', ax=axes[i], alpha=0.6,
                                     color='tomato', label='Churned', bins=30)
    axes[i].set_title(col)
    axes[i].legend()

plt.suptitle('Feature Distributions by Churn Status', fontsize=14, y=1.01)
plt.tight_layout()
plt.savefig('../screenshots/feature_distributions.png', dpi=150)
plt.show()
```

```python
plt.figure(figsize=(10, 7))
corr = df.select_dtypes(include=np.number).corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.savefig('../screenshots/correlation_heatmap.png', dpi=150)
plt.show()
```

```python
df.to_csv('../data/churn_cleaned.csv', index=False)
print("✅ Cleaned data saved!")
print("Shape:", df.shape)
print("Churn Rate:", round(df['Exited'].mean() * 100, 2), "%")
```
